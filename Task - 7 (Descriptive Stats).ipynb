Mean : It is a descriptive statistics which is used to find the average values.It is a type of Central tendency.
       It is also called a univariant.When the values of a data increases , the mean values get impacted. 
       
Formula : $\frac{\Sigma x_i}{n}$   where $x_i$ is each value of dataset ,  n=total number of values in dataset

```python
col=data['Annual_HH_Income']
# mean without library
mean= sum(col)/len(col)
print(mean)
plt.bar(1,mean)
```

```python
# mean with library , used the statistics library

print(st.mean(col))
```

Median : It is also a descriptive statistices which is used to find the middle values of the dataset.It is a type of Central tendency. It is also called univariant . If there is outlier in the dataset , the median value is robust to outlier.

n=number of values in dataset
First sort the data in the increasing order to find the median value.

if n is odd ,then 
median = ($\frac{n}{2}$)th term

if n is even, then median = (($\frac{n}{2}$)th term +($\frac{n}{2}$+1)th term)/2


```python
# median without library

col=list(col)
col.sort()
n=len(col)
if n%2==0:
    median= (col[n//2]+col[(n//2)-1])/2
else:
    median=col[n//2]
    
print(median)

plt.bar(1,median)
```

```python
# median with library

print(st.median(col))
```

Mode: It is a descriptive statistics which is used to find the number with maximum occurrence in the data set.It is a type of central tendency. It is robust to outlier.

Mode = maximum frequency of number

If we have 2 or more numbers with the same frequency then all the numbers are  said to be the modal values.

```python
# Mode without library

dic={}
for i in col:
    if i in dic:
        dic[i]+=1
    else:
        dic[i]=1
ma=0
for x in dic:
    if(ma<dic[x]):
        ma=dic[x]
        mode=x 
    
print("mode :",mode)
plt.bar(1,mode)
```

```python
# with library
print(st.mode(col))
```

Variance: It is a type of dispersion spread. It is a measure of variability. It is calculated by taking the squared deviation from mean. It also tells us the degree of spread in the dataset. The more the spread, the larger the variance.
It get impacted to outlier as mean is impacted to outlier. 
To remove outlier impact, we used squared deviation from median .                There are two types of variance :                                               i) Population variance :It is denoted by $\sigma^{2}$                           
ii) Sample variance : It is denoted by $S^{2}$                                  
Formula for variance:                                                                       
Population Variance($\sigma^{2}$) = $\frac{\Sigma_{i=1-n} (x_i-\mu)^{2}}{n}$   
Sample variance( $S^{2}$)= $\frac{\Sigma_{i=1-n} (x_i-\bar{x})^{2}}{n-1}$

```python
#variance without library
su=0
for i in col:
    su+= (i-mean)**2
pop_var=su/len(col)
sam_var=su/(len(col)-1)
print("Poulation Variance :",pop_var)
print("Sample Variance :",sam_var)
plt.bar(1,pop_var)
```

```python
# Variance with library

print(st.pvariance(col))
print(st.variance(col))
```

Standard Deviation : It is used to calculate the dispersion spread. It is nothing but the square root of Variance.                                                           
Population Standard Deviation ( $\sigma$) =$\sqrt[2]{PopulationVariance}$  ;     Sample Standard Deviation (S)= $\sqrt[2]{SampleVariance}$ 

```python
# Standard deviation without library
pop_std=pop_var**0.5
sam_std=sam_var**0.5

print("Population Standard Deviation :",pop_std)
print("Sample Standard Deviation :",sam_std)

plt.bar(1,pop_std)
```

```python
# standard Deviation using library

print("Population Standard Deviation :",st.pstdev(col))
print("Sample Standard Deviation :",st.stdev(col))
```

Normal Distribution : It is a probability distribution that is symmetric about the mean , showing that the data near the mean are in large number of occurrence than data far from the mean . It is also called Gaussian distribution.As the number of data increases ,it behaves like a normal distribution. It depends on mean and variance. 
  

```python
normal=np.random.normal(Y)
sns.distplot(normal)
```

Features of Normal Distribution:                                               

1) It is a bell curve shape figure i.e it is symmetric about the mean.          
2) It has the same mean, median and mode.                                        
3) It follows the 68%, 95%, 99.7% probability distribution rules. It means that  by  moving 1 stdev , 2 stdev , 3 stdev from mean to both the sides 
 

```python
normal_10000 = np.random.normal(loc = 20, scale = 5, size=15000)

sns.distplot(normal_10000)
```

```python
mean_normal= normal_10000.mean()

sigma = normal_10000.std()

one_std_right= mean_normal+sigma
one_std_left= mean_normal-sigma
two_std_right= mean_normal+(2*sigma)
two_std_left= mean_normal-(2*sigma)
three_std_right= mean_normal+(3*sigma)
three_std_left= mean_normal-(3*sigma)
```

```python
plt.figure(figsize=(20,10))
sns.set_style("darkgrid")
sns.distplot(normal_10000)

plt.axvline(mean_normal, color='coral', label='Mean')

plt.axvline(one_std_right, color='yellow', label='Mean + 1SD')
plt.axvline(one_std_left, color='yellow', label='Mean - 1SD')
plt.axvline(two_std_right, color='green', label='Mean + 2SD')
plt.axvline(two_std_left, color='green', label='Mean - 2SD')
plt.axvline(three_std_right, color='blue', label='Mean + 3SD')
plt.axvline(three_std_left, color='blue', label='Mean - 3SD')
plt.legend();
```

Positively Skewed Normal distribution :  It is a normal distribution in which  more data are clustered around the left tail side and the right tail are more flatten . It is also called right-skewed distribution. In this case mode< median < mean .                          

Negatively Skewed Normal distribution :It is a normal distribution in which  more data are clustered around the right tail side and left tail are more flatten . It is also called left-skewed distribution. In this case mean< median < mode .

Effect on mean , median and mode due to skewness:                                

** If skewness=0 , then all mean,median and mode are equal .                     

** If skewness>0 , then mode< median< mean  .

** If skewness<0 , then mean < median < mode  .
